# Zadanie 2: Perceptron dla zadania OR i XOR

```python
import torch
import torch.nn as nn
```

```python
x_OR = torch.tensor([[0,0],[0,1],[1,0],[1,1]], dtype=torch.float32)
y_OR = torch.tensor([0,1,1,1], dtype=torch.float32)

x_XOR = torch.tensor([[0,0],[0,1],[1,0],[1,1]], dtype=torch.float32)
y_XOR = torch.tensor([0,1,1,0], dtype=torch.float32)
```

```python
class Perceptron(nn.Module):
    def __init__(self):
        super(Perceptron, self).__init__()

        self.fc1 = nn.Linear(2,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, X):
        X = self.fc1(X)
        pred = self.sigmoid(X)
        return pred
```

```python
loss_fn = nn.BCELoss()
```

```python
model = Perceptron()
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
```

```python
n_iters = 1000

for epoch in range(n_iters):
    model.train()

    pred = model(x_OR)
    loss = loss_fn(pred.squeeze(), y_OR)
    
    loss.backward()
    optimizer.step()

    optimizer.zero_grad()
```

```python
to_predict = torch.tensor([[1,1]], dtype=torch.float32)
```

```python
with torch.inference_mode():
    model.eval()
    pred = model(to_predict)

```
