# Zadanie 3: Problem XOR

```python
import torch
import torch.nn as nn
```

```python
x_XOR = torch.tensor([[0,0],[0,1],[1,0],[1,1]], dtype=torch.float32)
y_XOR = torch.tensor([0,1,1,0], dtype=torch.float32)
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP,self).__init__()

        self.input_layer = nn.Linear(in_features=2,out_features=4)
        self.hidden_layer = nn.Linear(in_features=4,out_features=1)
        self.output = nn.Sigmoid()

    def forward(self,X):
        pred = self.output(self.hidden_layer(self.input_layer(X)))
        return pred
```

```python
model = MLP()
optimizer = torch.optim.Adam(params=model.parameters(),lr=0.1)
loss_fn = nn.BCELoss()
```

```python
n_epoch = 1000

for epoch in range(n_epoch):
    model.train()

    pred = model(x_XOR)
    loss = loss_fn(pred.squeeze(),y_XOR)

    loss.backward()
    optimizer.step()

    optimizer.zero_grad()
```

```python
to_predict = torch.tensor([[1,1]], dtype=torch.float32)
```

```python
with torch.inference_mode():
    model.eval()
    pred = model(to_predict)
```
