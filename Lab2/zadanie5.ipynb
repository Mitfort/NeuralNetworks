# Trening z Optimizer i Loss

```python
import torch
import torch.nn as nn
import torch.optim as optim

from tqdm import tqdm
```

```python
x = torch.randn((100,1), dtype=torch.float32)
noise = torch.rand((100,1))

y = 2 * x + 1 + noise
```

```python
y.shape
```

```python
w = torch.randn(1, dtype=torch.float32, requires_grad=True)
b = torch.randn(1, dtype=torch.float32, requires_grad=True)
```

```python
loss_fn = nn.MSELoss()
optimizer = optim.SGD(params=[w,b], lr=0.01)
```

```python
n_epochs = 1000
losses = []

for epoch in tqdm(range(n_epochs)):
    y_pred = x * w + b
    loss = loss_fn(y_pred, y)
    losses.append(loss.item())

    loss.backward()
    optimizer.step()

    optimizer.zero_grad()

    if epoch % 100 == 0:
        print(f"Epoch: {epoch} | Loss: {loss.item()}")

print(f"Final loss: {loss.item()}")
print(f"w: {w.item()}, b: {b.item()}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(losses)
plt.show()
```
